# AutoGrad

```python
x = torch.ones(2,2, requires_grad = True)
x
```

```python
y = x+2
y
```

```python
print(y.grad_fn)
```

```python
z = y*y*3
out = z.mean()
```

```python
print(z, out)
```

```python
out.backward()
```

```python
print(x.grad)
```

```python
print(x.requires_grad)
print((x**2).requires_grad)
```

```python
with torch.no_grad():
    print((x**2).requires_grad)
```
