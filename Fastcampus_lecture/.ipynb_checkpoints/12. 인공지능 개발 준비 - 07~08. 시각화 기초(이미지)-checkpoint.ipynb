## package 불러오기

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

%matplotlib inline
```

---

## 이미지 파일 열기

opencv로도 열수 있지만, shape 순서가 바뀔 때도 있어서 PIL로 여는 것을 개인적으로 선호

```python
path = 'source/dog.jpg'

image = Image.open(path)
image = np.array(image)
```

---

## 이미지 들여다 보기 

```python
image.shape
```

```python
np.min(image), np.max(image)
```

이미지를 열기 전에 shape 및 min, max를 통해서 이미지의 range 확인 필요 

___

## 그래프로 시각화 하기

```python
plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

---

## 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

---
## 이미지 흑백으로 열기

```python
image_bw = Image.open(path).convert("L")
image_bw = np.array(image_bw)
```

```python
image_bw.shape
```

```python
plt.imshow(image_bw)
plt.show()
```

---
### 다른 색상으로 cmap 표현하기

### gray scale

```python
plt.imshow(image_bw, 'gray')
plt.show()
```

---

### RdBu

```python
plt.imshow(image_bw, 'RdBu')
plt.show()
```

---

### jet

```python
plt.imshow(image_bw, 'jet')
plt.show()
```

---
## Colorbar 추가하기

```python
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```
