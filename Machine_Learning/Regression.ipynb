Linear Regression

```python
from sklearn import datasets
diabetes = datasets.load_diabetes() #당뇨병 데이터( 결과값 : 1년뒤 측정한 당뇨병의 진행률, 피처 : 10종)
print(diabetes.data.shape)
```

```python
num_test = 30
x_train = diabetes.data[:-num_test, :]
y_train = diabetes.target[:-num_test]

x_test = diabetes.data[-num_test:, :]
y_test = diabetes.target[-num_test:]
```

```python
from sklearn.linear_model import SGDRegressor

regressor = SGDRegressor(loss='squared_loss', penalty='l2', alpha=0.001, learning_rate='constant', eta0=0.01)
```

```python
regressor.fit(x_train,y_train)
predictions = regressor.predict(x_test)

print(predictions)
```

Decision Tree Regression

```python
from sklearn import datasets

boston = datasets.load_boston() # 결과값 : 1978 보스턴 주택 가격, 피처 : 13종
num_test = 10

boston.data.shape
```

```python
x_train = boston.data[:-num_test, :]
y_train = boston.target[:-num_test]

x_test = boston.data[-num_test:, :]
y_test = boston.target[-num_test:]
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(max_depth = 10, min_samples_split=3)
regressor.fit(x_train, y_train)

predictions = regressor.predict(x_test)
print(predictions)
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=100, max_depth=10, min_samples_split=3)

regressor.fit(x_train, y_train)
predictions = regressor.predict(x_test)
print(predictions)
```

Support Vector Regression
- classification과 다르게 regression 때는 hyperplane에 최대한 가깝게 만듦

```python
from sklearn.svm import SVR

regressor = SVR(C=0.1, epsilon=0.02, kernel='linear')
regressor.fit(x_train, y_train)
predictions = regressor.predict(x_test)

print(predictions)
```
