```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import  classification_report, confusion_matrix

# 데이터 셋 : https://www.kaggle.com/ronitf/heart-disease-uci
data = pd.read_csv('/content/drive/MyDrive/ML&DM/Machine_Learning/heart.csv')
```

```python
data = np.asarray(data)
x,y  = data[:,0:-1], data[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(y_test)
```

```python
parameters = {'hidden_layer_sizes' : [(50,50,50), (100, 20,10), (60,60,60) ], 'solver' : ['sgd', 'adam'], 'alpha' :[0.1, 0.01, 0.001], 'learning_rate': ['constant','adaptive'], 'max_iter' :[100,200,300,500,1000], 'learning_rate_init':[0.1, 0.01, 0.001, 0.0001]}

clf = MLPClassifier( random_state=1)
grid_search = GridSearchCV(clf, parameters, n_jobs=-1, cv=3)
grid_search.fit(x_train, y_train)
print(grid_search.best_params_)
```

```python
clf_best = grid_search.best_estimator_
clf_best.fit(x_train,y_train)
print("훈련 세트 정확도: {:.3f}".format(clf_best.score(x_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(clf_best.score(x_test, y_test)))
```

```python
pred = clf_best.predict(x_test)
pred
```

```python
print("\nClassification report:\n")
print(classification_report(y_test, pred))
```

```python
print("\nConfusion Matrix:\n")
tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
pd.DataFrame([[tn, fp], [fn, tp]], index = ["label: female(0)", "label: male(1)"], columns = ["predicted: female(0)", "predicted: male(1)"])
```
